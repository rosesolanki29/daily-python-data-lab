```python
# without parameter without return type
def sum():
    a=int(input("enter any number:"))
    b=int(input("enter any number:"))
    c=a+b
    print(c)
```

```python
sum()
```

```python
# with parameter without return type
def diff(a,b):
    c=a-b
    print(c)
```

```python
diff(10,2)
```

```python
n1=int(input("enter any number: "))
n2=int(input("enter any number: "))
diff(n1,n2)
```

```python
# without parameter with return type
def div():
    a=int(input("enter any number:"))
    b=int(input("enter any number:"))
    c=a/b
    return c
```

```python
div()
```

```python
print(div())
```

```python
# with parameter with return type
def mult(a,b):
    return a*b
```

```python
mult(10,2)
```

```python
# optional argument(parameter)
def mult1(a=10,b=2):
    return a*b
```

```python
mult1(5)
```

```python
def multi(a,b=5):
    return a*b
```

```python
multi(10,6)
```

```python

def sort_1(l,reverse=False):
    if reverse==True:
        for i in range(len(l)):
            for j in range(i+1,len(l)):
                if l[i]<l[j]:
                    l[i],l[j]=l[j],l[i]
    else:
        for i in range(len(l)):
            for j in range(i+1,len(l)):
                if l[i]>l[j]:
                    l[i],l[j]=l[j],l[i]
    return l
```

```python
sort_1([100,1,20,40],reverse=False)
```

```python
l=[100,1,23,45,21,2]
print(sort_1(l,reverse=True))
```
