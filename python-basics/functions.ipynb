```python

def sort_1(l,reverse=False):
    if reverse==True:
        for i in range(len(l)):
            for j in range(i+1,len(l)):
                if l[i]<l[j]:
                    l[i],l[j]=l[j],l[i]
    else:
        for i in range(len(l)):
            for j in range(i+1,len(l)):
                if l[i]>l[j]:
                    l[i],l[j]=l[j],l[i]
    return l
```

```python
sort_1([100,1,20,40],reverse=False)
```

```python
l=[100,1,23,45,21,2]
print(sort_1(l,reverse=True))
```

```python
def leap_year():
    year=int(input("enter any year:"))
    if year%4==0:
        if year%100==0:
            if year%400==0:
                print(year,"leap year")
            else:
                print(year,"not a leap year")
        else:
            print(year,"leap year")
    else:
        print(year,"leap year")
```

```python
leap_year()
```

```python
def leap_year1(year):
    if year%4==0:
        if year%100==0:
            if year%400==0:
                return True
            else:
                return False
        else:
            return True
    else:
        return False
```

```python
n=int(input("enter any year"))
if leap_year1(n):
    print(n,"leap year")
else:
    print(n,"not leap year")
```

```python
def prime(num):
    if num==0 or num==1:
        return False
    elif num>1:
        for i in range(2,num):
            if num%i==0:
                return False 
                break
        else:
            return True
    else:
        return False
```

```python
n=int(input("enter any number: "))
if prime(n):
    print(n,"prime number")
else:
    print(n,"not a prime")
```

```python
def prime1(num):
    if num==0 or num==1:
        return num,"not a prime"
    elif num>1:
        for i in range(2,num):
            if num%i==0:
                return num,"not a prime" 
                break
        else:
            return num,"prime"
    else:
        return num,"not a prime"
```

```python
prime1(7)
```

```python
def prime2(start,stop):
    l=[]
    for i in range(start,stop):
        if i>1:
            for j in range(2,i):
                if i%j==0:
                    break
            else:
                l.append(i)
    return l
```

```python
prime2(10,20)
```

```python
def fibo(n):
    a=0
    b=1
    if n==1:
        return a
    elif n==2:
        return b
    else:
        for i in range(2,n):
            c=a+b
            a=b
            b=c
    return c
```

```python
fibo(7)
```

```python
def fibo1(n):
    if n==1:
        return 0
    elif n==2:
        return 1
    fib=[0,1]
    for i in range(2,n):
        next_term=fib[-1]+fib[-2]
        fib.append(next_term)
    return fib[-1]
```

```python
fibo1(7)
```
