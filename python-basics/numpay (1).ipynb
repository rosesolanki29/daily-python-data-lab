```python
# indexcing 1d
arr=np.array([10,20,30,40,50])
print(arr)
print(arr[0])
print(arr[-1])
```

```python
#2D
arr=np.arange(1,13).reshape(3,4)
print(arr)
print(arr[1])
print(arr[1][2])
```

```python
#3D
arr=np.arange(1,28).reshape(3,3,3)
print(arr)
```

```python
# DIRECT ITERATION 
arr=np.array([10,20,30,40,50])
print(arr)
for i in arr:
    print(i)
```

```python
# indirect itrration
arr=np.array([10,20,30,40,50])
for i in range(len(arr)):
    print(arr[i])
```

```python
#2d 
arr=np.arange(1,13).reshape(3,4)
for i in arr:
    for j in i:
        print(j)
```

```python
arr=np.arange(1,28).reshape(3,3,3)
print(arr)
for i in arr:
    for j in i:
        for k in j:
            print(k)
```

```python
arr=np.arange(1,28).reshape(3,3,3)
for i in np.nditer(arr):
    print(i)
```

```python
arr=np.arange(1,28).reshape(3,3,3)
for i,j in np.ndenumerate(arr):
    print(i,j)
```
