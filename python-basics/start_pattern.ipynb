```python
# wap to print sum of n even number and n odd number and also count of each ? While loop
n=int(input("Enter a number: "))
even_sum=0
odd_sum=0
even_count=0
odd_count=0
i=1
while i<=n:
    if i%2==0:
        even_sum=even_sum+i
        even_count=even_count+1
    else:
        odd_sum=odd_sum+i
        odd_count=odd_count+1
    i=i+1
print("even sum:" ,even_sum,"even count: ",even_count)
print("odd sum:" ,odd_sum,"odd count: ",odd_count)
```
