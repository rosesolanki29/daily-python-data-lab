```python
#wap to check wheather a given year is leap year or not
#century_year% 400==0
#non-century_year%4==0
year=int(input("enter any number : "))
if year%4==0:
    if year%100==0:
        if year%400==0:
            print(year,"leap yar")              #century
        else:
            print(year,"not a leap year")       #century
    else:
        print(year,"leap year")                 #non-century
else:
    print(year,"not a leap year")               #non-century
            
```

```python
 #wap to check wheather a given triangle is valid or invalid if the triangle is valid check is equilateral
#scalen or isoselence (take 3 input)
side1=int(input("enter side: "))
side2=int(input("enter side: "))
side3=int(input("enter side: "))
if side1+side2>side3 and side1+side3>side2 and side2+side3>side1:
    if side1==side2==side3:
        print("equilateral triangle")
    elif side1==side2 or side2==side3 or side3==side1:
        print("isoselence")
    else:
        print("scalen")
else: 
    print("triangle is valid")
 
```
