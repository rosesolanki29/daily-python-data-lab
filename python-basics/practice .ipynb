```python
#Question 23. Hollow Square Pattern
#Accept a positive integer N and display a hollow square using ’*’characters.

n=5
for i in range(n):
    for j in range(n):
        if j==0 or i==n-1 or i==0 or j==n-1:
            print("*",end=" ")
        else:
            print(" ",end=" ")
    print()
```

```python
#Question 24. Water Tank Monitor

#A water tank has a maximum capacity of 1000 liters.
#Accept the current water level and display whether the tank is Empty,Half Filled, Almost Full, or Overflow.

water=int(input("enter number: "))
if water==0:
    print("empty tank")
if water<=400:
    print("half filled")
if water<=700:
    print("almost full")
if water<1000:
    print("overflow")

```

```python
# Question 25. Reverse Number Ladder

# Accept a positive integer N.
# For N = 5 display:

# 54321 
# 5432 
# 543 
# 54 
# 5

n = int(input("Enter any number: "))
for i in range(n,0,-1):
    for j in range(n,n-i,-1):
        print(j, end="")
    print()
```

```python
# Question 29. Right-Aligned Star Pattern

#         *
#       * *
#     * * *
#   * * * *
# * * * * *

# Accept a positive integer N and display a right-aligned triangle using
# ’*’ characters.

n=5
for i in range(n):
    for j in range(n-i-1):
        print(" ",end=" ")
    for j in range(i+1):
        print("*",end=" ")
    print()
```
