# Fibonacci Series

### In fibonacci series the next term is the sum of last 2 terms , which follows the formula , f(n) = f(n-1)+f(n-2)

```python
Term     = 1st 2nd 3rd 4th 5th 6th 7th --------------------- infinity
Series   =  0   1   1   2   3   5   7  --------------------- infinity  <--Real Series
Position =  0   1   2   3   4   5   6  --------------------- infinity
position is indexing in python
```

```python
# wap to print fibonacci series up to n'th term?
a = 0 
b = 1 
n = int(input("How Many Terms:-"))
print(a,b, end=" ")
for i in range (2,n):
    c = a+b
    print(c, end=" ")
    a = b
    b = c
```

```python
#wapp to check weather a given number is an arms strong number or not?
#arms strong number is a positive number equall to the sum of its own digit where each digit is raise to the power of the number of digit .

n=int(input("enter any number: "))
a=len(str(n))
sm=0
old=n
while n!=0:
    dig=n%10
    sm=sm+(dig**a)
    n=n//10
if sm==old:
    print(sm,"armstrong number")
else:
    print(sm,"not armstong number")
```
