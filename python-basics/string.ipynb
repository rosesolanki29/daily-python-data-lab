```python
#indexing & slicing

#index- position of character in a given string
#1st character of a string -- 0(index)
#last character of any string ---( -1)

#indexing:- is a technique for accesing indiviual unit for a given string

s="welcome to bhopal"
print(s[0])
print(s[-1])
print(s[4])
print(s[-3])
```

```python
#slicing:- technique for accessing multiple character from a given string 
#s[start:stop:step] 
#start - by default (0)
#stop - by default end of the string
# # if stop=n it will run to (n-1)
#step - by default +1

s="welcome to bhopal"
print(s[::])
print(s[::-1])
print(s[-2:-7:-1])
print(s[1:12:2])
print(s[:7:])
```
