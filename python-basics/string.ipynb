```python
#indexing & slicing

#index- position of character in a given string
#1st character of a string -- 0(index)
#last character of any string ---( -1)

#indexing:- is a technique for accesing indiviual character unit for a given string

s="welcome to bhopal"
print(s[0])
print(s[-1])
print(s[4])
print(s[-3])
```

```python
#slicing:- technique for accessing multiple character from a given string 
#s[start:stop:step] 
#start - by default (0)
#stop - by default end of the string
# # if stop=n it will run to (n-1)
#step - by default +1

s="welcome to bhopal"
print(s[::])
print(s[::-1])
print(s[-2:-7:-1])
print(s[1:12:2])
print(s[:7:])
```

```python
# wap program to print no of vowels, no of consonants and no of spaces in a given string?
s="welcome to city of lakes"
c_v=0
c_c=0
c_s=0
for i in s:
    if i in "aeiou":
        c_v=c_v+1
    elif i==" ":
        c_s=c_s+1
    else:
        c_c=c_c+1
print("count of vowels:",c_v)
print("count of consonants:",c_c)
print("count of spaces:",c_s)
```

```python
s="welcome to city of lakes"
c_v=0
c_c=0
c_s=0
for i in range(len(s)):
    if s[i] in "aeiou":
        c_v=c_v+1
    elif s[i]==" ":
        c_s=c_s+1
    else:
        c_c=c_c+1
print("count of vowels:",c_v)
print("count of consonants:",c_c)
print("count of spaces:",c_s)
```

```python
# index() - position of any character or substring from given string.
# strobj.index(char/substring,[start],[stop])
# by default- 0
# stop - by default - end of the string , n-1
s="welcome to bho bhopal"
print(s.index("o"))
print(s.index("o",5))
print(s.index("bhopal"))
print(s.index("o",5,9))
print(s.index("w"))
```

```python
# find() - position of any character or substring from given string.
# strobj.find(char/substring,[start],[stop])
# by default- 0
# stop - by default - end of the string , n-1
s="welcome to bho bhopal"
print(s.find("o"))
print(s.find("o",5))
print(s.find("bhopal"))
print(s.find("o",5,9))
print(s.find("w")) 
```

```python
# count() - occurence of any character/substring in a given string
# strobj.count(char/substring,[start],[stop])
s="welcome to city of lakes"
print(s.count("o"))
print(s.count("o",5))
print(s.count("o",5,9))
print(s.count("city"))
print(s.count(" "))
```
