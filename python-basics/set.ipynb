```python
# set - collections of elements
# denoted by {} with csv.
# store only unique value
# unordered,unindexed,mutable and heterogenous in nature.
s=set()   # empty set
print(s)
print(type(s))
```

```python
s={10,20,30,40,10,20,10,"hello",10,20,10.5}
print(s)
print(type(s))
s.add(1000)   # mutable
print(s)
```

```python
# iteration
s={10,20,30,40,10,20,10,"hello",10,20,10.5}
for i in s:
    print(i)
```

```python
# len()
s={10,20,30,40,"hello",10.5}
print(len(s))
count=0
for i in s:
    count=count+1
print(count)
```

```python
s={100,1,10.5,20,99,-10,91}
print(max(s))
print(min(s))
```

```python
# add element to an existing set
s={10,20,10,30,40,"hello",10.5,11}
print(s)
s.add(1000)
print(s)
```

```python
# pop()
s={10,20,30,40,"hello",10.5,100}
print(s.pop())
print(s)
```

```python
# remove()
s={10,20,30,40,"hello",10.5}
print(s)
s.remove(30)
print(s)
s.remove(1000)
print(s)
```

```python
# discard()
s={10,20,30,40,"hello",10.5}
print(s)
s.discard(30)
print(s)
s.discard(1000)
print(s)
```

```python
s1={100,20,200,30,50}
s2={200,40,60,50,400,20}
print(s1.union(s2))
print(s1.intersection(s2))
print(s1.difference(s2))
print(s1.symmetric_difference(s2))
```

```python
s1={100,20,200,30,50}
s2={200,40,60,50,400,20}
print(s1|s2)
print(s1&s2)
print(s1-s2)
print(s1^s2)
```

```python
# frozenset() -collection of unique elements
# denoted by - constructor --- frozenset()
# unordered,unindexed and immutable and heterogenous
f=frozenset({10,20,10,30,40,"hello",10.5,10,10,20})
print(f)
print(type(f))
```

```python
f=frozenset()
print(f)
print(type(f))
```
