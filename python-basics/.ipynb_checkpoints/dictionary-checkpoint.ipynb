```python
# dictionary 
# collection of elements denoted by {k:v}
# keys must be unique and values can be duplicate
# ordered,mutable and unindexed
d={}   # empty dictionary 
print(d)
print(type(d))
```

```python
d={"name":["abc","def","pqr"],"subject":["DA","DE","DS"],"fees":[30000,45000,40000]}
print(d)
print(type(d))
print(d["name"])    # unindexed
d["name"]="decode"  # keys --unique and mutable
print(d)
d["Subject"]="java"
print(d)
```

```python
d={}
d["name"]="pqr"
print(d)
```

```python
l=[1,2,3,4,5]
# {1:1,2:8,3:27,4:64,5:125}
d={}
for i in l:
    d[i]=i**3
print(d)
```

```python
l1=[5,1,2,3,4,6]
l2=[3,6,5,4,2,1]
#{5:125,1:1,2:32,3:81,4:16,6:6}
d={}
for i,j in zip(l1,l2):
    d[i]=i**j
print(d)
```

```python
#{"A":1,"B":4,"C":9,"D":16}
d={}
for i in range(1,5):
    d[chr(64+i)]=i**2
print(d)
```

```python
l=["red","green","red","blue","red","blue","Yellow","yellow"]
#{"red":3,"green":1,"blue":2,"Yellow":1,"yellow":1}
d={}
for i in l:
    if i in d:
        d[i]+=1
    else:
        d[i]=1
print(d)
```

```python
# dictionary iteration 
# direct
d={'name': ['abc', 'def', 'pqr'], 'subject': ['DA', 'DE', 'DS'], 'fees': [30000, 45000, 40000]}
for i in d:
    print(i,d[i])
```

```python
d={'name': ['abc', 'def', 'pqr'], 'subject': ['DA', 'DE', 'DS'], 'fees': [30000, 45000, 40000]}
for i in d:
    for j in d[i]:
        print(j)
```
