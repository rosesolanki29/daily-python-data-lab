```python
import numpy as np
```

```python
age=np.array([20,21,20,23,24,26,23,24,34,23,21,27,28,28,24,23,25,29,26,27,28,29,23,24,25,
              26,23,31,28,29,12])
print(age)
```

```python
print(np.mean(age))
print(np.median(age))
```

```python
sum=0
count=0
for i in age:
    count=count+1
    sum=sum+i
mean=sum/count
print(mean)
```

```python
min=np.quantile(age,0)
print(min)
```

```python
q1=np.quantile(age,0.25)
print(q1)
```

```python
q2=np.quantile(age,0.50)
print(q2)
```

```python
q3=np.quantile(age,0.75)
print(q3)
```

```python
max=np.quantile(age,1)
print(max)
```

```python
IQR=q3-q1
print(IQR)
```

```python
lower_fence=q1-(1.5*IQR)
print(lower_fence)
```

```python
upper_fence=q3+(1.5*IQR)
print(upper_fence)
```

```python
l=[]
for i in age:
    if i<lower_fence or i>upper_fence:
        l.append(i)
print(l)
print(len(l))
```
