# operator

## Arithmetic Operator

```python
a= 10
b=2
print("addition:",a+b)
print("difference:",a-b)
print("multiplication:", a*b)
print("simple divison :",a/b)
print("floor division:",a//b)
print("modulus:",a%b)
print("exponential power :",a**b)
```

```python
#assignment operator =,+=,-=,*=,//=,%=,**=
a=10
b=2
c=a+b  #12
a+=b  #a=a+b  #a=12
b*=5  #b=b*5 #b=10
c/=2 #c=c/2  #6.0
c=a+b   #c=22
print(a,b,c)
```

```python
#comparision operator
a=10
b=8
print(a==b)
print(a!=b)
print(a>=b)
print(a<=b)
print(a>b)
print(a<b)
```

```python
#logical operator -and or not
a=10
b=8
c=12
print(a>b and b>c)
print(a>b or b>c)
print(a>b and not b>c)
```

```python
#identity operator is
#id() - that return memory  address of a given object
a=10
b=8
print(id(a))
print(id(b))
print( a is b)
print( a is not b)
```

```python
a=10
b=10
print(id(a))
print(id(b))
```

```python
#membership operator is not in 
a="hello"
print("h" in a)
print("h" not in a)
print("H" in a)  # python - case sensitive 
```

```python
#bitwise operator &
a=10
b=8
print(bin(a))
print(bin(b))
print()
print(bin(a&b))
print()
print(a&b)
```

```python
#bitwise operator or |- pipe
a=10
b=8
print(bin(a))
print(bin(b))
print()
print(bin(a|b))
print()
print(a|b)
```

```python
a=10
b=8
print(bin(a))
print(bin(b))
print()
print(bin(a^b)) 
print()
print(a^b)
```
