```python
#numeric datatype
#integer
#type-return class(datatype) of any given object
a=10
print(a)
print(type(a))


```

```python
#float
a=10.5
print(a)
print(type(a))
```

```python
#string - ordered sequence of character
#single line string denoted by '' or " "
#multi-line string denoted by ''' '''
#indexed,immutable
a= ''        #empty string
print(a)
print(type(a))
```

```python
s="hello"
print(s)
print(type(s))
print(s[0])    #indexed
s[0]="z"    # immutable
print(s)
```

```python
s= ''' hello
welcome to city of lakes '''
print(s)
print(type(s))
```

```python
#list-is a collection of elements
#denoted by[] with csv
#ordered, indexed,mutable and hetrogenous in nature
l=[] #empty list
print(l)
print(type(l)) 
```

```python
l=[10,20,30,"hello",10.5]
print(l)
print(type(l))
print(l[3])   #indexed
l[3]="decode"   #mutable
print(l)

```

```python
#tuple- is collection of elements
#denoted by () with csv
#ordered,indexed,immutable and heterogenous in nature
t=() #emty tuple
print(t)
print(type(t))
```

```python
#single valued tuple
t=(10,)
print(t)
print(type(t))
```

```python
t=(10,20,30,"hello",10.5)
print(t)
print(type(t))
print(t[2])  #indexed
t[2]=3000
print(t)
```

```python
#dictionary- collection of element
#denoted by {k:v}
#keys must be unique,values can be duplicated
#unindexed,ordered,mutable
d={}   #emty dectionary
print(d)
print(type(d))

```

```python
d={"name":"abc","course":"DA","fees":25000}
print(d)
print(type(d))
print(d["name"])  #unindexed
d["name"]="pqr"  #mutable
print(d)
   
```

```python
d={"name":["abc","pqr","def"],"course":["DA","DE","DS"],"fees":[250000,23000,30000]}
print(d)
d["name"][-1]="decode"
print(d)
```

```python
d={}
d["name"]="decode"
print(d)
```

```python
#set - collection of element
#denoted by {} with csv
#stored unique values only
#unordered,unindexed,mutable and hetrogenous in nature
s=set() #set constructor
print(s)
print(type(s))
```

```python
s={10,20,30,"hello",40,10.5,20,10,10,20}
print(s)
print(type(s))
s.add(1000)   #mutable
print(s)
print(s[0]) #unindexed
```
