```python
class vehical:
    def __init__(self,name="Thar",model="2022"): # this is a constructor. 
        self.name = name
        self.model = model
        
    def display(self): # this is method 
        print(f"vehical name is {self.name} and model is {self.model}")

        
    @staticmethod
    def addition(a,b):
        print("addition of number",a+b)

v1 = vehical()
print(v1.display())
print(v1.addition(12,34))
        

v2 = vehical("dzire","2022")
print(v2.display())

v3 = vehical("farari","2026")
print(v3.display())
```

```python
class vehical1:
    class_name = "this is a global variable"  # class varable 

    def __init__(self,name,model):
        self.name = name
        self.model = model  # 

    def display(self):
        print(f"vehical name is {self.name} and model is {self.model}")

    @classmethod
    def class_variable(cls):   # class method 
        print(f"class variable name is {cls.class_name}")

    @staticmethod
    def addition(a,b):
        print(f"addition of {a+b}")     # local function 

v1 = vehical1("dzire","2022")
print(v1.display())
print(v1.class_variable())
print(v1.addition(10,20))
```

```python
# public variable  (self.name = name)
# private variable (self.__name = name)
# protected varialble (self._name = name)

class person:
    def __init__(self,name,salary):
        self.__name = name   # private varialble 
        self.salary = salary # public variable

    def getdetails(self):
        print(f"Name is {self.__name} and salary is {self.salary}")

p1 = person("gagan",20000)
print(p1.getdetails())

p2 = person("samshad",25000)
print(p2.getdetails())

print(p2.salary)

        
```

```python
# basic getter and setter method
class account:
    def __init__(self,amount):
        self.__amount= amount

    # getter methode
    def getamount(self):
        print(f"amount is: {self.__amount} ")

    # setter methode
    def setamount(self,setAmt):
        self.__amount = setAmt
        print(f"amount after set new amount: {self.__amount}")

acc1 = account(1000)
print(acc1.getamount())
print(acc1.setamount(25000))

```

```python
# setter getter metode by the use of decorators
# property -- it helps us to get details of private variabble 


class account:
    def __init__(self,amount):
        self.__amount= amount

    @property # getter methode
    def amount(self):
        print(f"amount is: {self.__amount} ")

    @amount.setter # setter methode
    def amount(self,setAmt):
        self.__amount = setAmt
        print(f"amount after set new amount: {self.__amount}")

acc1 = account(1000)
print(acc1.amount)
acc1.amount = 25000
print(acc1.amount)
```

```python
# hw 
class employees:
    def __init__(self,name , age , salary):
        self.name = name
        self.age = age
        self.__salary  = salary 

    @property
    def salary(self):  # getter method
        return self.__salary

    @salary.setter
    def salary(self,salary):   # setter method
        self.__salary = salary 

    def details(self):
        return (f"employee name : {self.name}\n"
                f"employee age : {self.age}\n"
                f"employee salary : {self.__salary}"
                )


emp = employees("sam",20,150000)
print(emp.details())

emp.salary = 250000
print("new salary : ",emp.salary)

        
```

```python
# inheritance 
# it is property that allows one class to aquire attribute and method of another class.
# single inheritance
class parent:
    def displayA(self):
        print("this is a parent class")

class child(parent):
    def displayB(self):
        print("this is a child class")

obj = child()
obj.displayA()

```

```python
# 2 multilevel inheritance
class A:
    def displayA(self):
        print("class A")

class B(A):
    def displayB(self):
        print("class B")

class C(B):
    def displayC(self):
        print("class c")

obj = C()
obj.displayA()
obj.displayB()

obj2 = B()
obj2.displayA()
    
```

```python
# 3 multiple inheritance   \

class A:
    def displayA(self):
        print("class A")

class B:
    def displayB(self):
        print("class B")

class C(A,B):
    def displayC(self):
        print("class c")

obj = C()
obj.displayA()
obj.displayB()
```

```python
# 4 heirarichal inheritance

class A:
    def displayA(self):
        print("class A")

class B(A):
    def displayB(self):
        print("class b")

class C(A):
    def displayC(self):
        print("class C")

obj = B()
obj.displayA()

obj = C()
obj.displayA()

```

```python
# hybride inheritance .. 
class A:
    def displayA(self):
        print("class a")
class B(A):
    def displayB(self):
        print("class B")
class C(A):
    def displayC(self):
        print("class C")

class D(B,C):
    def displayD(self):
        print("class D")

class D2(A):
    def displayD2(self):
        print("class d2")

obj = D()
obj.displayA()
obj.displayB()
obj.displayC()
obj.displayD()

obj2 = D2()
obj2.displayA()
obj2.displayD2()
```

```python
# parent constructor call by child use of super()

class parent:
    def __init__(self,name):
        self.name = name
        print("parent constructor called")

class child(parent):
    def __init__(self,name,age):
        super().__init__(name)
        self.age = age
        print("child constructor called")

obj = child("sam",20)
print(obj.__dict__)
        
```

```python
# MRO -- method resolution order -- searching method if method name is same... 
class A:
    def display(self):
        print("class a")

class B(A):
    def display(self):
        print("class B")

class C(A):
    def display(self):
        print("class C")

class D(B,C):
    pass 

obj = D()
obj.display()

print(A.mro())
print(B.mro())
print(C.mro())
print(D.mro())    
```

```python
# polymorphism... --> polymorphism is opps allows same method or operator behaves differently 
# according to data .. 

# 1. overrriding 
# method overrriding 

# 2. overloading 
# method overloading 
# operator overloading 


# methode overriding .. 
class animal():
    def speak(self):
        print("animal can sound : ")
        
class dog(animal):
    def speak(self):
        print("dog is bark")

class cat(animal):
    def speak(self):
        print("cat speaks meow ")

d = dog()
c = cat()
d.speak()
c.speak()
```

```python
# method overloading.. --> python doesnot support direct method overloading.. 
class calculation():
    def addition(self,*number):  # we can add infinit numbers without specific arguments.. *args , **wargs 
        return sum(number)

obj = calculation()
print(obj.addition(2,4))
print(obj.addition(2,5,6))
print(obj.addition(2,3,4,5,65))

```

```python
# operator overloading.. 
class employees():
    def __init__(self,salary):
        self.salary = salary 

    def __add__(self,other):   # magic function  or dunder method 
        return self.salary + other.salary

e1 = employees(5000)
e2 = employees(3000)

e1+e2

#   """
  #  __sub__ substraction
   # __gt__ greaterthan
   # __et__ equal overloading
   # __mul__ multiplication overloading 
    
 #   """

 
```

```python
# abstraction -->  in abstraction we hide implementatiton details and show only necessary 
# data /method to users .. 

# abstract also use to inforce rules to child class; 

# ABC - abstract base class

from abc import ABC,abstractmethod

class bankAccount(ABC):
    @abstractmethod
    def withdraw(self,withdraw_amount):
        pass 

    @abstractmethod
    def deposit(self,deposit_amount):
        pass

class sbi(bankAccount):
    def withdraw(self):
        return "this is abstract method"

    def deposit(self):
        return "this is abstrat class deposit method"
        
obj  = sbi()
obj.withdraw()
```
