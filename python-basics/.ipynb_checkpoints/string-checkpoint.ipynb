```python
#indexing & slicing

#index- position of character in a given string
#1st character of a string -- 0(index)
#last character of any string ---( -1)

#indexing:- is a technique for accesing indiviual character unit for a given string

s="welcome to bhopal"
print(s[0])
print(s[-1])
print(s[4])
print(s[-3])
```

```python
#slicing:- technique for accessing multiple character from a given string 
#s[start:stop:step] 
#start - by default (0)
#stop - by default end of the string
# # if stop=n it will run to (n-1)
#step - by default +1

s="welcome to bhopal"
print(s[::])
print(s[::-1])
print(s[-2:-7:-1])
print(s[1:12:2])
print(s[:7:])
```

```python
#string- ordered sequence of character
# denoted by '' or ""
# immutable,indexed,iterable.
s=""   # empty string
print(s)
print(type(s))
```

```python
s="hello"
print(s)
print(type(s))
print(s[0])   # indexed
s[0]="z"  # immutable
print(s)
```

```python
# iteration
# direct iteration and indirect iteration
s="hello"
for i in s:
    print(i)
```

```python
# indirect iteration - with the help of range function
# len() - predefined function that return no of characters in a given string.
s="hello"
print(len(s))
```

```python
s="hello"
for i in range(len(s)):
    print(i,s[i])
```

```python
# wap program to print no of vowels, no of consonants and no of spaces in a given string?
s="welcome to city of lakes"
c_v=0
c_c=0
c_s=0
for i in s:
    if i in "aeiou":
        c_v=c_v+1
    elif i==" ":
        c_s=c_s+1
    else:
        c_c=c_c+1
print("count of vowels:",c_v)
print("count of consonants:",c_c)
print("count of spaces:",c_s)
```

```python
s="welcome to city of lakes"
c_v=0
c_c=0
c_s=0
for i in range(len(s)):
    if s[i] in "aeiou":
        c_v=c_v+1
    elif s[i]==" ":
        c_s=c_s+1
    else:
        c_c=c_c+1
print("count of vowels:",c_v)
print("count of consonants:",c_c)
print("count of spaces:",c_s)
```
