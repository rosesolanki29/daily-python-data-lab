```python
#indexing & slicing

#index- position of character in a given string
#1st character of a string -- 0(index)
#last character of any string ---( -1)

#indexing:- is a technique for accesing indiviual character unit for a given string

s="welcome to bhopal"
print(s[0])
print(s[-1])
print(s[-3])
```

```python
#slicing:- technique for accessing multiple character from a given string 
#s[start:stop:step] 
#start - by default (0)
#stop - by default end of the string
# # if stop=n it will run to (n-1)
#step - by default +1


s="welcome to bhopal"
print(s[: :])
print(s[:7:])
print(s[1:11:2])
print(s[-1:-10:-2])
print(s[: :-1])
print(s[-2:-12:-1])
```

```python
#string - ordered sequence of character
#denoted by '' or " "
#indexed, immutable
s="hello"
print(s)
print(type(s))
print(s[1])   #indexed
s[0]="z"      #immutable
print(s)
```

```python
# iteration
# a)- direct iteration 
#b)-indirect iteration
s="hello"
for i in s:
    print(i)
```

```python
# indirect iteration

s="hello"
for i in range(len(s)):
    print(i,s[i])
```

```python
#wapp to print number of vowels consonant and the spaces in a given ?
s="welcome to city of lakes"
no_c=0
no_v=0
no_s=0
for i in s:
    if i in "AEIOUaeiou":
        no_v=no_v+1
    elif i==" ":
        no_s=no_s+1
    else:
        no_c=no_c+1
print("number of vowels:",no_v)
print("number of consonants:",no_c)
print("number of space:",no_s)
```

```python
#wapp to print reverse of any given string
s=input("enter any string: ")
print(s[::-1])
```

```python
s=input("enter any string: ")
rev=""
for i in s:
    rev=i+rev
print(rev)
```

```python
#wapp to check weather a given string is palindrome or not
s=input("enter any string:")
rev=""
for i in s:
    rev=i+rev
if s==rev:
    print(s,"palindrome")
else:
    print(s,"not a palindrome")
```

```python
s=input("enter any string:")
if s==s[::-1]:
    print(s,"palindrome")
else:
    print(s,"not a palindrome")
```

```python
#functions
#len()- return no of characters in a given string
#len(strobj)
s="hello"
print(len(s))
count=0
for i in s:
    count=count+1
print(count)
```
