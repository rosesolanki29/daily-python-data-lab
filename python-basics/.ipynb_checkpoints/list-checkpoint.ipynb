```python
#list-collection of elements
#denoted by [] with csv
#ordered,indexed,mutable and hetrogenous
l=[]
print(l)
print(type(l))
```

```python
l=[10,20,30,"hello",200.5]
print(l)
print(type(l))
print(l[0])    #indexed
l[0]=100       #mutable
print(l)
```

```python
#iteration
#direct iteration
l=[10,20,30,"hello",200.5]
for i in l:
    print(i)
```

```python
#indirect iteration----with help of range function
l=[10,20,30,"hello",200.5]
for i in range (len(l)):
    print(i,l[i])
```

```python
#indexing
l=[10,20,30,40,"hello",200,300]
print(l[0])
print(l[-1])
print(l[4][-1])
```

```python
l=[10,20,30,[100,200,["bpl","ind"],10.5,20.5],1000,2000]
print(l)
print(l[3][2][1])
```

```python
#slicing
l=[10,20,30,40,"hello",200,300]
print(l[::])
print(l[::-1])
print(l[1::2])
print(l[:5:])
```

```python
#functions
#len()- no of elements in a given list
l=[10,20,30,"decode",100.5]
print(len(l))
count=0
for i in l:
    count=count+1
print(count)
```

```python
#max()
#min()
l=[10,20,30,99,100.5]
print(max(l))
print(min(l))
```

```python
l=[10,192,34,65]
print(max(l))
max_value=l[0]
for i in l:
    if i>max_value:
        max_value=i
print(max_value)
```

```python
l=[10,34,83,56,97,182]
sum=0
for i in l:
    sum=sum+i
print(sum)
```

```python
l=[10,34,83,56,97,182]
sum=0
for i in range(len(l)):
    sum=sum+l[i]
print(sum)
```

```python
l=[7,8,7,2,3,3,7,8,7,7,7,3,5,4,7]
sum=0
count=0
for i in l:
    sum=sum+i
    count=count+1
print(sum)
print(count)
print(sum/count)

    
```

```python
l=[7,8,7,2,3,3,7,8,7,7,7,3,5,4,7]
if len(l)%2==0:
    print(l[len(l)/2]+l[len(l)])
else:
    print(l[len(l)//2])
```
