```python
#wap to print count of even and odd number between 2000 and 3000
start=int(input("enter any number: "))
stop=int(input("enter any number: "))
count_even=0
count_odd=0
for i in range(start,stop+1):
    if i%2==0:
        count_even=count_even+1
    else:
        count_odd=count_odd+1
print("even number between",start,"and",stop,":",count_even)
print("odd number between",start,"and",stop,":",count_odd)
```

```python
#wapp to check wheather a given number is prime or not?
n=int(input("enter any number: "))
if n==0 or n==1:
    print(n,"not a prime")
elif n>1:
    for i in range(2,n):
        if n%i==0:
            print(n,"not a prime")
            break
    else:
        print(n,"prime number")
else:
    print(n,"not a prime")
            
    
```

```python
# wapp to print factorial of number range between (5 and 15)
start=int(input("enter any number: "))
stop=int(input("enter any number: "))
for i in range(start,stop+1):
    fact=1
    for j in range(1,i+1):
        fact=fact*j
    print(i,"-",fact)
```

```python
#wapp to print sum & count of N even number & n odd number between 1000 and 2000
start=int(input("enter any number: "))
stop=int(input("enter any number: "))
even_sum=0
odd_sum=0
even_count=0
odd_count=0
for i in range(start,stop+1):
    if i%2==0:
        even_sum=even_sum+i
        even_count=even_count+1
    else:
        odd_sum=odd_sum+i
        odd_count=odd_count+1
print("even count: ",even_count)
print("even sum: ",even_sum)
print("odd count: ",odd_count)
print("odd sum: ",odd_sum)
```
