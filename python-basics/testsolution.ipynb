```python
#Write a Python program that determines the type of a triangle based on the lengths of its three sides. The types are: 
#Equilateral (all sides are equal) 
#Isosceles (two sides are equal) 
#Scalene (all sides are different) 
#Invalid (the sides do not form a valid triangle)

side1=int(input("enter side: "))
side2=int(input("enter side: "))
side3=int(input("enter side: "))
if side1+side2>side3 and side1+side3>side2 and side2+side3>side1:
    if side1==side2==side3:
        print("equilateral triangle")
    elif side1==side2 or side2==side3 or side3==side1:
        print("isoselence")
    else:
        print("scalen")
else: 
    print("triangle is valid")
 
```

```python
#Write a Python program that takes a year as input and checks if it is a leap year. 
#If it is a leap year, print "Leap year," otherwise print "Not a leap year." 
year=int(input("enter year: "))
if year%4==0:
    if year%100==0:
        if year%400==0:
            print("leap year")
        else:
            print("not a leap year")      
    else:
        print(year,"leap year")                
else:
    print(year,"not a leap year")             
            
        
```

```python
#.Write a Python program that takes an integer as input and determines whether it is positive, negative, or zero. 
#Additionally, if the number is positive, check if it is even or odd. 
num=int(input("enter any number: "))
if num>0:
    if num%2==0:
        print("positive and even number")
    else:
        print("odd")
elif num<0:
    print("negative")
else:
    print("zero")
    
```

```python
# Eligibility for Special Programs 
#Objective: Write a Python program that determines whether a student is eligible for an honors program or needs 
#remedial classes based on their performance in different subjects. 
#Details: Eligibility criteria: 
#Honors: Average score of 85 or above across all subjects and no subject below 70. 
#Remedial: Any subject below 50. 
#Input:-Scores in three subjects: Math, Science, and Language (out of 100) Output: 
#"Eligible for Honors" 
#"Needs Remedial" 
#"Standard Placement"

maths=int(input("enter any number: "))
science=int(input("enter any number: "))
language=int(input("enter any number: "))
avg_marks=(maths+science+language/3)
if avg_marks>85 and maths>=70 and science>=70 and language>=70:
    print("honors")
elif maths<50 or science<50 or language<50:
    print("remidial")
else:
    print("standard placement")

```
