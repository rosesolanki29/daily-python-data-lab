```python
import random
l=("S","P","R")
user=input("For Rock - R, For Paper - P, For Scissor - S :- ")
comp=random.choice(l)
print("Computer :- ",comp)
if user in l:
    if user==comp:
        print("DRAW")
    elif (user=='R' and comp=='S') or (user=='P' and comp=="R") or (user=="S" and comp=='P'):
        print("Winner 🔥🔥🔥🔥🔥")
    else:
        print("Loser 💀💀💀💀💀")
else:
    print("Invalid Output")
```
