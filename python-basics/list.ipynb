```python
#iteration
#direct iteration
l=[10,20,30,"hello",200.5]
for i in l:
    print(i)
```

```python
#indirect iteration----with help of range function
l=[10,20,30,"hello",200.5]
for i in range (len(l)):
    print(i,l[i])
```

```python
l=["madam","bhopal","decode","aa","Naman","nitin","111"]
l1=[]
for i in l:
    rev=""
    for j in i:
        rev=j+rev
    if i==rev and len(i)>3:
           l1.append(i)
print(l1)
```
