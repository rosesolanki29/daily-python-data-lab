```python
#sorting

#sorted()                                                     #sort()
#-->  sorted is a function                             --> sort is a method
#-->  always return new list                           --> always return existing listobj
#-->  sorted(seqobj,[revers=false])                    -->l.sort([reverse=false])
#       by default return false                              by default false
```

```python
#sorted()--- function
#always return new sorted list
#sorted(seqobj,[reverse=false])
#by default ascending order
l=[10,1,34,5]
l1=sorted(l)
print(l1)

l2=sorted(l,reverse=True)
print(l2)
```

```python
#sort()-- method
#existing listobj
#l.sort([reverse=false])
l=[10,1,34,5]
l.sort()
print(l)

l2.sort(reverse=True)
print(l2)
```

```python
l=[1,10,9,8,7,2,2,2,4,5,5,7,7,7,9,11,3,5,7,10]
sum=0
count=0
for i in l:
    sum=sum+i
    count=count+1
mean=sum/count
print(mean)
sorted_list=sorted(l)
n=len(l)
if n%2!=0:
    median=sorted_list[n//2]
else:
    median=(sorted_list[n//2]+sorted_list[n//2-1])/2
print(median)
```

```python
l1=[5,1,2,3,6,4]
l2=[2,6,5,4,1,3]
#[25,1,32,81,6,64]
l=[]
for i,j in zip(l1,l2):
    l.append(i**j)
print(l)
print([i**j for i,j in zip(l1,l2)])
```

```python
#enumerate(listobj)-return position and elements.
l=[10,20,30,40,50,60,70]
for i,j in enumerate(l):
    print(i,j)  #i- elements position j- elemnts
```

```python
#list comprehension- concise way to create a list
#syntax-[expression looping statements conditional statements]
l=[1,2,3,4,5]
#[1,8,27,64,125]
l1=[]
for i in l:
    l1.append(i**3)
print(l1)
print([i**3 for i in l])
```

```python
l=[1,4,5,6,2,3,9]
#[1,125,27,729]
l1=[]
for i in l:
    if i%2!=0:
        l1.append(i**3)
print(l1)
print([i**3 for i in l if i%2!=0])
print([l[i]**3 for i in range(len(l)) if l[i]%2!=0])
```
