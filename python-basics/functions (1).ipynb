```python

def sort_1(l,reverse=False):
    if reverse==True:
        for i in range(len(l)):
            for j in range(i+1,len(l)):
                if l[i]<l[j]:
                    l[i],l[j]=l[j],l[i]
    else:
        for i in range(len(l)):
            for j in range(i+1,len(l)):
                if l[i]>l[j]:
                    l[i],l[j]=l[j],l[i]
    return l
```

```python
sort_1([100,1,20,40],reverse=False)
```

```python
l=[100,1,23,45,21,2]
print(sort_1(l,reverse=True))
```

```python
def leap_year():
    year=int(input("enter any year:"))
    if year%4==0:
        if year%100==0:
            if year%400==0:
                print(year,"leap year")
            else:
                print(year,"not a leap year")
        else:
            print(year,"leap year")
    else:
        print(year,"leap year")
```

```python
leap_year()
```

```python
def prime(num):
    if num==0 or num==1:
        return False
    elif num>1:
        for i in range(2,num):
            if num%i==0:
                return False 
                break
        else:
            return True
    else:
        return False
```

```python
n=int(input("enter any number: "))
if prime(n):
    print(n,"prime number")
else:
    print(n,"not a prime")
```

```python
def prime2(start,stop):
    l=[]
    for i in range(start,stop):
        if i>1:
            for j in range(2,i):
                if i%j==0:
                    break
            else:
                l.append(i)
    return l
```

```python
def arm(start,stop):
    l=[]
    for i in range(start,stop+1):
        a=len(str(i))
        sumdigit=0
        old=i
        while i!=0:
            dig=i%10
            sumdigit=sumdigit+(dig**a)
            i=i//10
        if sumdigit==old:
            l.append(sumdigit)
    return l
```

```python
# wap function that return fibonacci series upto nth term?
def fibo(n):
    if n==1:
        return [0]
    elif n==2:
        return [0,1]
    fib=[0,1]
    for i in range(2,n):
        next_term=fib[-1]+fib[-2]
        fib.append(next_term)
    return fib
```

```python
fibo(7)
```

```python
def fibo1(n):
    if n==1:
        return 0
    elif n==2:
        return 1
    a,b=0,1
    for i in range(2,n):
        c=a+b
        a,b=b,c
    return c   
```

```python
#wapf that return largest number amoung 3 no.
def max_of_two(a,b):
    if a>b:
        return a
    else:
        return b
```

```python
def max_of_three(x,y,z):
    return max_of_two(x,max_of_two(y,z))
```

```python
max_of_three(20,30,100)
```

```python
def max_of_three_num(a,b,c):
    if a>b and a>c:
        return a
    elif b>a and b>c:
        return b
    else:
        return c
```

```python
def power (base,exp):
    if exp==0:
        return 1
    else:
        return base*power(base,exp-1)
```

```python
#wap recursive function that return nth term of fibonacci?
def fib(n):
    if n==1:
        return 0
    elif n==2:
        return 1
    else:
        return fib(n-1)+fib(n-2)
```

```python
def rev(n):
    rev_num = 0
    while n != 0:
        dig = n % 10
        rev_num = rev_num * 10 + dig
        n = n // 10
    return rev_num
```
